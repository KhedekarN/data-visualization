## Finding missing values

```python
df.info()
```

    two methods to find missing values: isna and isnull 
    both are exactly same methods with the different names

```python
# o/p is pandas series with T/f values
df["Interest_Paid"].isna().head()
```

```python
interest_missing= df["Interest_Paid"].isna()
```

```python
#Looks at the row with missing values
df.loc[interest_missing,:]
```

```python
# using ~ to negate filter so that will have results without missing values
df.loc[~interest_missing,:]
```

```python
df["Interest_Paid"].isna().sum()
```

```python
df.info()
```

## Removing missing values

you can remove missing values using dropna method

```python
# here we are dropping row if they contain any or all nans but this is not best strategy
df[30:40].dropna(how='any')
```

## Filling missing values
There are different ways of filling missing values

```python
#finding where missing value is located 
df["Interest_Paid"][30:40]
```

```python
# filling zero as nan but its not a good idea
df["Interest_Paid"][30:40].fillna(0)
```

```python
#backfill in value
df["Interest_Paid"][30:40].fillna(method='bfill')
```

```python
#forwardfill in value
df["Interest_Paid"][30:40].fillna(method='ffill')
```

```python
#Linear Interpolation(fill in of vlaues)
df["Interest_Paid"][30:40].interpolate(method='linear')
```

```python
df["Interest_Paid"].sum()
```

```python
interest_missing=df["Interest_Paid"].isna()
df.loc[interest_missing,"Interest_Paid"] = 93.24
```

```python
df['Interest_Paid'].sum()
```

```python
#we don't have missing values in df dataframe anymore
df.info()
```
