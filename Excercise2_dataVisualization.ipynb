# MATLAB-STYLE

```python

plt.plot(month_no,interest_paid,c='k')
plt.plot(month_no,principal_paid,c='b')
```

```python
plt.plot(month_no,interest_paid,c='k',label='interest')
plt.plot(month_no,principal_paid,c='b',label='principal')
#set xlim and ylim
plt.xlim(left=1,right=70)
plt.ylim(bottom=0,top=1000)
#set labels
plt.xlabel('months',fontsize=15)
plt.ylabel('money paid',fontsize=15)
#setting title
plt.title('Interest and pricipal paid every month',fontsize=25)
#changing tick font size
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#to have grid line
#plt.grid()
# only verticle gridlines
#plt.grid(axis='y')
# only horizontal gridlines
plt.grid(axis='x')
plt.grid(c='r',
        alpha=0.8,
        linestyle='--')
plt.grid(axis='x')
#setting legend
#plt.legend(loc='best',fontsize=15)
plt.legend(loc=(0.3,1.08),borderaxespad=0,fontsize=15)
# saving file 
plt.tight_layout()
plt.savefig('C:/Users/nikit/Desktop/Ex_Files_Python_Data_Visualization/Exercise_Files/3 Pandas/data/matlabstyle.png',dpi=300)
```

# Object-Oriented Style

```python
fig,axes=plt.subplots(nrows=1,ncols=1)
axes.plot(month_no,interest_paid,c='k',label='interest')
axes.plot(month_no,principal_paid,c='b',label='principal')
#setting xlim,ylim
axes.set_xlim(left=1,right=70)
axes.set_ylim(bottom=0,top=1000)
#setting the labels
axes.set_xlabel("months",fontsize=20)
axes.set_ylabel("money",fontsize=20)
#setting titles
axes.set_title("Interest and pricipal paid every month",fontsize=25)
#Setting tick fontsize
axes.tick_params(axis='x',labelsize=13)
axes.tick_params(axis='y',labelsize=13)
#grids
#axes.grid()
#horizontal
#axes.gris(axis='x')
#verticle
axes.grid(axis='y')
axes.grid(c='r',
        alpha=0.8,
        linestyle='--')
axes.grid(axis='y')
#setting legend
#axes.legend(loc='best',fontsize=15)
axes.legend(loc=(0.3,1.08),borderaxespad=0,fontsize=15)
#saving as a fig
fig.tight_layout()
fig.savefig('C:/Users/nikit/Desktop/Ex_Files_Python_Data_Visualization/Exercise_Files/3 Pandas/data/oostyle.png',dpi=300)
```
